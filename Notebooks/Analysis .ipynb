```python
#Import necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('../Data/MpesaData.csv')
df.head()
```

```python
#Cleaning the dataset
df.info()
df.describe()
```

```python
#Financial Inclusion Analysis
plt.figure()
plt.plot(df["Year"], df["Financial_Inclusion_Percent"])
plt.title("Financial Inclusion in Kenya (2006–2024)")
plt.xlabel("Year")
plt.ylabel("Percent")
plt.grid()
plt.savefig("../Visuals/inclusion_trend.png")
plt.show()
```
